### Test Environments

```python
environment_name="Breakout-v0"
```

```python
env=gym.make(environment_name)
```

```python
print(env.action_space.sample())
print(env.action_space)
# print(env.observation_space.sample())
print(env.observation_space)
```

```python
episodes = 5
for episode in range(1, episodes+1):
    state=env.reset()
    done=False
    score=0
    
    while not done:
        env.render()
        action=env.action_space.sample()
        n_state,reward,done,info=env.step(action)
        score+=reward
    print(f"Episode :{episode} Score : {score}")

env.close()
```

```python
env.close()
```

### Vectorise Environment and Train Model

```python
# Vectorizing the environment, 
# particularly with multiple environments, 
# allows you to train the agent faster by training in parallel
```

```python
env=make_atari_env("Breakout-v0",n_envs=4,seed=0)
env=VecFrameStack(env,n_stack=4)
```

```python
env.reset()
```

```python
env.render() #4 environments
```

```python
env.close()
```

```python
#setting up model for training
log_path=os.path.join("Training","logs")
model=A2C("CnnPolicy", env, verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=40000)
```

```python
# after 40000 tiemsteps, average reward mean is 4.37
```

### Save and Reload Model

```python
a2c_path=os.path.join("Training","Saved Models", "A2c_model")
```

```python
model.save(a2c_path)
```

```python
del model
```

```python
model=A2C.load(a2c_path,env)
```

### Evaluate and Test

```python
evaluate_policy(model, env, n_eval_episodes=10,render=True)
```

```python
# avg rewards(score), stadard deviation
```

```python
obs=env.reset()
score=0

while True:
    action,_=model.predict(obs)
    obs,rewards, dones, info=env.step(action)
    env.render()
    score+=rewards
    print(f"Score : {score}")
```
