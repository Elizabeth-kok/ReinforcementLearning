```python
import os
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

## Load Environment

```python
environment_name="CartPole-v0"
```

```python
# creating the environment
env=gym.make(environment_name)
```

```python
# testing out the environment
episodes=5
for episode in range(1,episodes+1): # 1, 2, 3, 4, 5
    state=env.reset() # we get the initial set of observations
    done=False
    score=0
    
    while not done:
        env.render() # allows us to view the graphical representation of the environment
        action=env.action_space.sample() # generating a random action
        n_state,reward,done,info= env.step(action)
        score+=reward
    print(f"Episode :{episode} Score :{score}")
```

```python
env.close()
```

```python
state=env.reset() # we get the initial set of observations
action=env.action_space.sample()
n_state,reward,done,info= env.step(action)

print("state : " ,state) # 4 observations
print("action : ",action) # action is either 1 or 0
print("n_state,reward,done,info : ",n_state,reward,done,info)
```

## Understand the Environment

```python
# 0-push cart to the left , 1-push cart to the right
env.action_space.sample()
```

```python
#[cart position, cart velocity, pole angle, pole angular velocity]
env.observation_space.sample()
```

## Train a RL model

```python
log_path=os.path.join("Training","Logs")
```

```python
env=gym.make(environment_name)
env=DummyVecEnv([lambda:env])
model=PPO("MlpPolicy", env, verbose=1, tensorboard_log=log_path) # defining the agent
```

```python
env
```

```python
model
```

```python
#train the model
model.learn(total_timesteps=10000)
```

## Save and Reload Model

```python
PP0_path=os.path.join("Training","Saved Models","PP0_model")
```

```python
model.save(PP0_path)
```

```python
del model
```

```python
model=PPO.load(PP0_path,env=env)
model
```

## Evaluation

```python
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True) #(avg score, standard deviation?)
```

## Test Model

```python
episodes=5
for episode in range(1,episodes+1): # 1, 2, 3, 4, 5
    obs=env.reset() # we get the initial set of observations
    done=False
    score=0
    
    while not done:
        env.render() # allows us to view the graphical representation of the environment
        action,_=model.predict(obs) # generating a random action
        obs,reward,done,info= env.step(action)
        score+=reward
    print(f"Episode :{episode} Score :{score}")
```

## Adding a calback to the training stage

```python
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```python
save_path=os.path.join("Training", "Saved Models")
log_path=os.path.join("Training","Logs")
```

```python
env=gym.make(environment_name)
env=DummyVecEnv([lambda:env])
```

```python
stop_callback=StopTrainingOnRewardThreshold(reward_threshold=190,verbose=1)
eval_callback=EvalCallback(env,
                          callback_on_new_best=stop_callback,
                           eval_freq=10000,
                           best_model_save_path=save_path,
                           verbose=1)
```

```python
model=PPO("MlpPolicy", env, verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=20000,callback=eval_callback)
```

```python
model_path=os.path.join("Training","Saved Models","best_model")
model=PPO.load(model_path,env=env)
```

```python
evaluate_policy(model, env, n_eval_episodes=10 ,render=True)
```

```python
env.close()
```

## Changing Policies

```python
#modifying neural network architecture
# changing the no of units and layers inside of your neural network (custom actor, value function)
net_arch=[dict(pi=[128,128,128,128], vf=[128,128,128,128])]
net_arch
```

```python
model=PPO("MlpPolicy", env, verbose=1, policy_kwargs={"net_arch":net_arch})
```

```python
model.learn(total_timesteps=20000,callback=eval_callback)
```
